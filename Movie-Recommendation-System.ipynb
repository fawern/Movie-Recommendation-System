```python
import numpy as np 
import pandas as pd 

import warnings 
warnings.filterwarnings('ignore')
```

```python
data = pd.merge(
    pd.read_csv('./Data/Dataset.csv'), 
    pd.read_csv('./Data/Movie_Id_Titles.csv'),
    on='item_id'
)

df = data.copy()

df.drop('timestamp',axis=1, inplace=True)

df.head(10)
```

```python
top_ratings = pd.DataFrame(df.groupby('title')['rating'].mean()).sort_values(by='rating')

top_ratings['most_ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())

top_ratings.sort_values(by='most_ratings', ascending=False).head(10)
```

```python
movies = df.pivot_table(index='user_id', columns='title', values='rating')

movies.head(10).T
```

```python
sample_movie = movies['Star Wars (1977)']
sample_movies_with = movies.corrwith(sample_movie)

corr_df = pd.DataFrame(sample_movies_with, columns=['Correlations'])

print('Total null values:', corr_df.isnull().sum().values)

corr_df.dropna(inplace=True)
print('Total null values:',corr_df.isnull().sum().values)
```

```python
similar_movies = corr_df[corr_df['Correlations'].values >= 0.5]

similar_movies = similar_movies.join(top_ratings['most_ratings'])

similar_movies[similar_movies.most_ratings >= 10].sort_values('Correlations', ascending=False)
```
